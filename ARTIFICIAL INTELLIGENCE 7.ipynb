```python
from imageai.Detection import ObjectDetection
detector = ObjectDetection()
```

pada task ini akan menjalankan yolo tiny dan akan mengiport 1.jpg lalu setelah itu akan di keluarkan dengan format newimage.jpg

```python
model_path = "yolo-tiny.h5"
input_path = "./image/1.jpg"
output_path = "newimage.jpg"
```

dana kan di scan dengan program yolo tiny v3

```python
detector.setModelTypeAsTinyYOLOv3()
```

```python
detector.setModelPath(model_path)
```

```python
detector.loadModel()
```

```python
detections = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
```

```python
for eachitem in detections:
  print(eachitem["name"] , " : ", eachitem["percentage_probability"])
```
